```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.cluster import KMeans
import pickle
df=pd.read_csv("song_data.csv")
df_1=df
# df['cluster']=0
df.tail()
```

```python
clmns=['song_popularity','acousticness','danceability','energy','tempo','liveness','speechiness']
X = stats.zscore(df[clmns])
```

```python
kmeans=KMeans(n_clusters=10)
kmeans.fit(X)
filep=open("KM.pkl","wb")
pickle.dump(kmeans,filep)
filep.close()
```
