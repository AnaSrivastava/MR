```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.cluster import KMeans
import pickle
df=pd.read_csv("song_data.csv")
df_1=df
df['cluster']=0
df.tail()
```

```python
clmns=['song_popularity','acousticness','danceability','energy','tempo','liveness','speechiness']
X = stats.zscore(df_1[clmns])
```

```python
kmeans=KMeans(n_clusters=10)
kmeans.fit(X)
```

```python
# print(kmeans.cluster_centers_)
```

```python
# prediction = kmeans.predict([[73,0.005520,0.496,0.682,1,167.060,0.0589,0.0294]])
# print(prediction)
```

```python
# print(df_1[clmns].groupby(['clusters']).mean())
labels = list(kmeans.labels_)
for i in range(0,len(df)):
    print(labels[i])
    df.set_value(i, 'cluster', labels[i])
```
